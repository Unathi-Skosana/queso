```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pathlib
from matplotlib import cm

from queso.utils.io import IO
```

```python
io = IO(folder=r"2023-01-31_noisy_cfi_batch_sweep")
p = io.path.glob('**/*')
dfs = [io.load_dataframe(x.stem) for x in p if x.is_file()]
df = pd.concat(dfs)
```

```python
print(df.n.unique())
```

```python
#
fig, ax = plt.subplots()
gamma = 0.0
# for n in sorted(df.n.unique()):
for n in [2, 3, 4]:
    dfi = df[(df.n == n) & (df.gamma == gamma)]
    ax.scatter(dfi.k, dfi.cfi/n**2, label=f"n={n}")
ax.legend()
```

```python
#
fig, ax = plt.subplots()
k = 2
for n in sorted(df.n.unique()):
    dfi = df[(df.n == n) & (df.k == k)]
    ax.scatter(dfi.gamma, dfi.cfi/n**2, label=f"n={n}")
ax.legend()
```

```python
fig, ax = plt.subplots()
n = 2
for k in sorted(df.k.unique()):
    dfi = df[(df.n == n) & (df.k == k)]
    ax.scatter(dfi.gamma, dfi.cfi/n**2, label=f"n={n}")
ax.legend()
```
